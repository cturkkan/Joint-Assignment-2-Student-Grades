Simple Linear Regression with Using Scikit - Learn

```python
X = d3[["age"]]
y = d3[["G3_y"]]
```

Model

```python
reg_model = LinearRegression().fit(X, y)
```

y-had = b+ w+x

(b - bias)

```python
reg_model.intercept_[0]
```

Coefficent

```python
reg_model.coef_[0][0]
```

Prediction Process

```python
reg_model.intercept_[0] + reg_model.coef_[0][0] *18
```

Visualization of the Model

```python
g = sns.regplot(x=X, y=y, scatter_kws={'color': 'b', 's': 9}, ci=False, color="r")
g.set_title (f"Model Equlation: G3_Y = {round(reg_model.intercept_[0],2)} + Age *{round(reg_model.coef_[0][0],2)}")
g.set_ylabel("G3_Y")
g.set_xlabel("Age")
plt.xlim(15,22)
plt.ylim(0,20)
plt.show()
```

Prediction Success

I find the mean squared error by giving the predicted values and the actual values to mean squared error.

```python
y_pred = reg_model.predict(X)
mean_squared_error(y,y_pred)
```
