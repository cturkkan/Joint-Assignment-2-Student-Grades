Multiple Linear Regression

```python
numeric_data = d3.select_dtypes(include=['float', 'int'])
```

```python
X = numeric_data.drop('G3_y', axis=1)
y = numeric_data[["G3_y"]]
```

Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
reg_model = LinearRegression().fit(X_train, y_train)
```

```python
reg_model.intercept_[0]
```

```python
reg_model.coef_[0]
```


MULTIPLE REGRESSION MODEL

This is our train fault

```python
y_pred = reg_model.predict(X_train)
```

```python
np.sqrt(mean_squared_error(y_train, y_pred))
```

```python
Train RKare
```

```python
reg_model.score(X_train, y_train)
```

Test RMSE

We ask the model for the independent values â€‹â€‹of the test variable.

```python
y_pred = reg_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

The percentage of the independent variable influencing the dependent variable is 76%

```python
reg_model.score(X_test, y_test)
```

Average of 10-fold RMSE values

```python
np.mean(np.sqrt(-cross_val_score(reg_model, X, y, cv=10, scoring="neg_mean_squared_error")))
```
